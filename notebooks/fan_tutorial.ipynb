In symbolic computation, it's quite common to manipulate a complex tree structure, it would be nice that we can have a simple syntax to manipulate the tree structure. For example, suppose we have an abstract syntax as follows:

First, we turn *fan* mode on:

```ocaml
#fan;;
```

```ocaml
(* Abstract syntax for the arith DDSL *)
type arith =  
   [ `Add    of (arith * arith)
   | `Minus  of (arith * arith)
   | `Times  of (arith * arith)
   | `Divide of (arith * arith)
   | `Power  of (arith * arith)
   | `Let    of (string * arith * arith) (* let x = arith1 in arith2 *)
   | `Int    of int
   | `Var    of string]
```

Since we want to write a program analyzer over *arith* type, we want to simplify this syntax to minimal. In *Fan*, langauges are exensible and composable. Different languages are dlimited by a delimiter `%language_name{` and `}`. Fan has built-in languages, and users can add new language to Fan. The cool thing in Fan is that langauges are composable and nestable, it's very easy to compose different langauges.

Now suppose we want to define our own new tiny language *arith*.



The first thing is to write your tokenizer, cool thing is that Fan has a built in DDSL, *lex_fan*, very easy to write a tokenizer in *fan*

```ocaml
let rec token = %lex_fan{
 | @whitespace %{token lexbuf}
 | @ocaml_num_literal
 | @ocaml_lid("let"|"be"|"in")
 | @ocaml_ant
 | @kwd_symbol("+"|"-"|"*"|"/"|"**"|"("|")")
 | @ocaml_eof
 | @default};;
 
```

Note *token* accept *Lexing.lexbuf* as input, we want adapt it into a stream:

```ocaml
let lexer = Lexing_util.adapt_to_stream token;;
```

Now we want to create a *grammar* for lexer. For this simple example, we will merge *parsing* and *compilation* in one step::

```ocaml
%create{ arith };;
```

*create* is anoter built-in delimited domain specific language, *Fan* also has built-in delimited domain specific langauge for parsing::

```ocaml
let add_binop left nt action prec op = %extend{
    nt: $prec $bool:left
      [ nt; $key:op; nt   ${action}]}
```

Function *add_binop* shows that Fan has a programmable parser, it defines a grammar over non-terminal *nt*, action *action*, precedence *prec* and operator *op*

```ocaml

 
let mk_action tag e2 _ e1 _loc =
  %ep{($vrn:tag ($e1,$e2) : arith)} 
 
let tbl = [("+",  ("Add",   10, true));
           ("-",  ("Minus", 10, true));
           ("*",  ("Times", 20, true));
           ("/",  ("Divide",20, true));
           ("**", ("Power", 30, false))] ;;
 

 
```

Now we start to populate *arith* grammar::

```ocaml
List.iter
 (fun (op,(tag,prec,left))
   -> add_binop left arith (mk_action tag) prec op) tbl ;;
```

We can visualize the generated grammar on the fly::

```ocaml
Gramf.dump Format.std_formatter arith
```

Now we define the grammar for *terminals*


```ocaml
%extend{
  arith: 40 true
    [ "("; arith as e; ")" %{e}
    | Lid s                %ep{`Var $str:s}
    | Int i                %ep{`Int $int:i} ]} ;;
 
```

```ocaml
Gramf.dump Format.std_formatter arith
```

We also want to support *atinquotaion* in *arith* DDSL.

```ocaml
let mk_ant nt prec = %extend{
    nt: $prec
      [ Ant("",x)   %{Tokenf.ant_expand Parsef.ep x}]}
```

```ocaml
mk_ant arith 40 ;;
```

```ocaml
Gramf.dump Format.std_formatter arith;;
```

```ocaml
%extend{
  arith: 0 false
    [ "let"; Lid x; "be"; arith as e1; "in"; arith as e2
        %ep{`Let($str:x,$e1,$e2)} ]} ;;
```

```ocaml
Gramf.dump Format.std_formatter arith;;
```

Finally, let's register our new language::

```ocaml
%register{ name : arith; position: ep; entry: arith; lexer: lexer}
```

```ocaml
let a = %arith{ 3 + 2 }
```

```ocaml
let b = function | %arith{ $x + 0 } -> x | y -> y
```

```ocaml
let c = %arith{let x be 3 in y}
```
